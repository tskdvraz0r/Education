# Step 04 - Сдвиг в развитии

На вход программе подается строка текста из натуральных чисел. Из элементов строки формируется список чисел. Напишите программу циклического сдвига элементов списка направо, когда последний элемент становится первым, а остальные сдвигаются на одну позицию вперед, в сторону увеличения индексов.

**Формат входных данных:** На вход программе подается строка текста из разделенных пробелами натуральных чисел.

**Формат выходных данных:** Программа должна вывести элементы измененного списка с циклическим сдвигом, разделяя его элементы одним пробелом.

## test: <br/>
Sample Input: <br/>
489 483 43 2 3 84 1 4 3 2 5 4 3 13

Sample Output: <br/>
13 489 483 43 2 3 84 1 4 3 2 5 4 3

```python
# 01.04.2022

num_str = input().split()

num_list = []
num_list.append(num_str[-1])

for i in range(len(num_str) - 1):
    num_list.append(num_str[i])

print(*num_list)
```
