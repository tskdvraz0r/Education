# Step 11 - Роскомнадзор запретил букву а 🌶️🌶️

Необходимо написать программу, реализующую алгоритм написания этой песни. Алгоритм выводит в конце предложения следующую в алфавитном порядке букву, если она встречается в строке текста, а очередную строку отображает уже без этой буквы.

**Формат входных данных:** На вход программе подается одно слово, записанное строчными русскими буквами без буквы "ё".

**Формат выходных данных:** Программа должна вывести в соответствии с указанным алгоритмом строки, количество которых равно количеству разных букв в строке, которая получается путем конкатенации введенного слова и строки "запретил букву".

## test: <br/>
Sample Input: <br/>
роскомнадзор <br/>

Sample Output: <br/>
роскомнадзор запретил букву а <br/>
роскомндзор зпретил букву б <br/>
роскомндзор зпретил укву в <br/>
роскомндзор зпретил уку д <br/>
роскомнзор зпретил уку е <br/>
роскомнзор зпртил уку з <br/>
роскомнор пртил уку и <br/>
роскомнор пртл уку к <br/>
росомнор пртл уу л <br/>
росомнор прт уу м <br/>
росонор прт уу н <br/>
росоор прт уу о <br/>
рср прт уу п <br/>
рср рт уу р <br/>
с т уу с <br/>
т уу т <br/>
уу у <br/>

```python
# 20.02.2022

abc = ('абвгдежзийклмнопрстуфхцчшщъыьэюя')
string = input() + ' запретил букву'

while len(string) > 0:
    for i in abc:
        if i in string:
            print(string + ' ' + i)
            string = string.replace(i, '').replace('  ', ' ').strip()
```
