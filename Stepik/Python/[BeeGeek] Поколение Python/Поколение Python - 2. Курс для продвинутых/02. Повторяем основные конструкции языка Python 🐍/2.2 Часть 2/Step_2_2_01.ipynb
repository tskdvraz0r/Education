# Step 01 - Координатные четверти

Дан набор точек на координатной плоскости. Необходимо подсчитать и вывести количество точек, лежащих в каждой координатной четверти.

**Формат входных данных:** В первой строке записано количество точек. Каждая следующая строка состоит из двух целых чисел — координат точки (сначала абсцисса – x, затем ордината – y), разделенных символом пробела.

**Формат выходных данных:** Программа должна вывести количество точек, лежащих в каждой координатной четверти, как в примерах.

**Примечание:** Учтите, что точки, лежащие на осях координат, не принято относить к какой-либо координатной четверти.

## test: <br/>
Sample Input: <br/>
10 <br/>
4 8 <br/>
-3 -1 <br/>
-4 9 <br/>
4 0 <br/>
-4 0 <br/>
5 -2 <br/>
0 0 <br/>
1 1 <br/>
13 -3 <br/>
-43 3 <br/>

Sample Output: <br/>
Первая четверть: 2 <br/>
Вторая четверть: 2 <br/>
Третья четверть: 1 <br/>
Четвертая четверть: 2 <br/>

```python
# 20.02.2022

n = int(input())
first_quarter = 0
second_quarter = 0
third_quarter = 0
fourth_quarter = 0

for i in range(n):
    coordinates = input().split(' ')
    if int(coordinates[0]) > 0 and int(coordinates[1]) > 0:
        first_quarter += 1
    if int(coordinates[0]) < 0 and int(coordinates[1]) > 0:
        second_quarter += 1
    if int(coordinates[0]) < 0 and int(coordinates[1]) < 0:
        third_quarter += 1
    if int(coordinates[0]) > 0 and int(coordinates[1]) < 0:
        fourth_quarter += 1

print(
    f'Первая четверть: {first_quarter}',
    f'Вторая четверть: {second_quarter}',
    f'Третья четверть: {third_quarter}',
    f'Четвертая четверть: {fourth_quarter}',
    sep = '\n'
)
```
