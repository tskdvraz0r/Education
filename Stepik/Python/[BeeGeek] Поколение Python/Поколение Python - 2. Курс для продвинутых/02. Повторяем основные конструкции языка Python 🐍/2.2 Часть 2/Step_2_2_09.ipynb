# Step 09 - Орел и решка

Дана строка текста, состоящая из букв русского алфавита "О" и "Р". Буква "О" – соответствует выпадению Орла, а буква "Р" – соответствует выпадению Решки. Напишите программу, которая подсчитывает наибольшее количество подряд выпавших Решек.

**Формат входных данных:** На вход программе подается строка текста, состоящая из букв русского алфавита "О" и "Р".

**Формат выходных данных:** Программа должна вывести наибольшее количество подряд выпавших Решек.

**Примечание:** Если выпавших Решек нет, то необходимо вывести число 0.

## test: <br/>
Sample Input: <br/>
ОРРОРОРООРРРО

Sample Output: <br/>
3

```python
# 01.05.2022

string = input().split("О")

res = 0
for i in range(len(string)):
    if len(string[i]) > res:
        res = len(string[i])
    else:
        continue

print(res)
```
