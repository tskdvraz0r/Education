# Step 02 - Больше предыдущего

На вход программе подается строка текста из натуральных чисел. Из неё формируется список чисел. Напишите программу подсчета количества чисел, которые больше предшествующего им в этом списке числа, то есть, стоят вслед за меньшим числом. 

**Формат входных данных:** На вход программе подается строка текста из разделенных пробелами натуральных чисел.

**Формат выходных данных:** Программа должна вывести одно число – количество элементов списка, больших предыдущего.

## test: <br/>
Sample Input: <br/>
1 2 3 4 5

Sample Output: <br/>
4

```python
# 20.02.2022

n = input().split(' ')
count = 0

for i in range(len(n) - 1):
    if int(n[i + 1]) > int(n[i]):
        count += 1

print(count)
```
