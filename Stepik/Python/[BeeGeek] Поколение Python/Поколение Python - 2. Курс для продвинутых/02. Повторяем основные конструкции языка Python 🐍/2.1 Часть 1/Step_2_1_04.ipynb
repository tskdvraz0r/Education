# Step 04 - Стоимость строки

Дана строка текста. Напишите программу для подсчета стоимости строки, исходя из того, что один любой символ (в том числе пробел) стоит 60 копеек. Ответ дайте в рублях и копейках в соответствии с примерами.

**Формат входных данных:** На вход программе подается строка текста.

**Формат выходных данных:** Программа должна вывести стоимость строки.

## test: <br/>
Sample Input: <br/>
Я собираюсь сделать ему предложение, от которого он не сможет отказаться.

Sample Output: <br/>
43 р. 80 коп.

```python
# 01.04.2022

string = input()
print(f'{(len(string) * 60) // 100} р. {(len(string) * 60) % 100} коп.')
```
