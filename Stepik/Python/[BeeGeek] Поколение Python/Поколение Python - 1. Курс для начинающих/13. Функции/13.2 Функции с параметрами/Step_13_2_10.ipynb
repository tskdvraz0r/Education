# Step 10 - Сумма цифр

Напишите функцию print_digit_sum(), которая принимает одно целое число num и выводит на печать сумму его цифр.

## test: <br/>
Sample Input: <br/>
12345

Sample Output: <br/>
15

```python
# 11.11.2021

def print_digit_sum(num):
    num = list(str(num))
    total = 0
    for i in num:
        total += int(i)
    print(total)

# считываем данные
n = int(input())

# вызываем функцию
print_digit_sum(n)
```
