# Step 08 - Звездный треугольник

Напишите функцию draw_triangle(fill, base), которая принимает два параметра:
- fill – символ заполнитель;
- base – величина основания равнобедренного треугольника;

а затем выводит его.

**Примечание:** Гарантируется, что основание треугольника – нечетное число.

## test: <br/>
Sample Input: <br/>
'*' <br/>
9 <br/>

Sample Output: <br/>
'*' <br/>
'**' <br/>
'***' <br/>
'****' <br/>
'*****' <br/>
'****' <br/>
'***' <br/>
'**' <br/>
'*' <br/>

```python
# 11.11.2021

def draw_triangle(fill, base):
    import math
    base_for = math.ceil(base / 2)
    
    for i in range(base + 1):
        if i < base_for:
            print(fill * i)
            
        elif i == base_for:
            print(fill * base_for)
        
        elif i > base_for:
            print(fill * (base_for - 1))
            base_for = base_for - 1

# считываем данные
fill = input()
base = int(input())

# вызываем функцию
draw_triangle(fill, base)
```
