Напишите функцию get_factors(num), принимающую в качестве аргумента натуральное число и возвращающую список всех делителей данного числа.

## test: <br/>
Sample Input: <br/>
5

Sample Output: <br/>
[1, 5]

```python
# 12.11.2021

def get_factors(num):
    result = []
    for i in range(1, num + 1):
        if num % i == 0:
            result.append(i)
    return result

# считываем данные
n = int(input())

# вызываем функцию
print(get_factors(n))
```
