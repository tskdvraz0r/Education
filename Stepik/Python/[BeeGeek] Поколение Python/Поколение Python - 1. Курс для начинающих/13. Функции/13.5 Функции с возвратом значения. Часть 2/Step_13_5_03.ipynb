# Step 03 - Is a Number Prime? 🌶️

Напишите функцию is_prime(num), которая принимает в качестве аргумента натуральное число и возвращает значение True если число является простым и False в противном случае.

## test: <br/>
Sample Input: <br/>
10

Sample Output: <br/>
False

```python
# 12.11.2021

# объявление функции
def is_prime(num):
    cnt = 0
    for i in range(n):
        if n % (i + 1) == 0:
            cnt += 1
            
    if cnt == 2:
        return True
    else:
        return False
    
# считываем данные
n = int(input())

# вызываем функцию
print(is_prime(n))
```
