# Step 10 - Змеиный регистр

Напишите функцию convert_to_python_case(text), которая принимает в качестве аргумента строку в «верблюжьем регистре» и преобразует его в «змеиный регистр».

## test: <br/>
Sample Input: <br/>
ThisIsCamelCased

Sample Output: <br/>
this_is_camel_cased

```python
# 13.11.2021

# объявление функции
def convert_to_python_case(text):
    for i in range(len(text)):
        if text[i].isupper():
            symb = '_' + text[i].lower()
            text = text.replace(text[i], symb)
    
    text = text.replace('_', '', 1)
    return text

# считываем данные
txt = input()

# вызываем функцию
print(convert_to_python_case(txt))
```
