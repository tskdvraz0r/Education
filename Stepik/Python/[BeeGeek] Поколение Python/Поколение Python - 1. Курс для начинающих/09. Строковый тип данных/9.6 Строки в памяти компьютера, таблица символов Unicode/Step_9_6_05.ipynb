# Step 05 - Простой шифр

На вход программе подается строка текста. Напишите программу, которая переводит каждый ее символ в соответствующий ему код из таблицы символов Unicode.

**Формат входных данных:** На вход программе подается строка текста.

**Формат выходных данных:** Программа должна вывести кодовые значения символов строки разделенных одним символом пробела.

## test: <br/>
Sample Input: <br/>
Hello world!

Sample Output: <br/>
72 101 108 108 111 32 119 111 114 108 100 33

```python
# 29.10.2021

s = input()
for i in range(len(s)):
    print(ord(s[i]), end=' ')
```
