# Step 12 - Сколько раз?

На вход программе подается одна строка. Напишите программу, которая определяет сколько раз в строке встречаются символы + и *.

**Формат входных данных:** На вход программе подается одна строка.

**Формат выходных данных:** Программа должна вывести сколько раз встречаются символы  + и * в строке.

## test: <br/>
Sample Input: <br/>
bcd+a++++**31415 <br/>

Sample Output: <br/>
Символ + встречается 5 раз <br/>
Символ * встречается 2 раз <br/>

```python
# 27.10.2021

string = input()
count_mult = 0
count_plus = 0

for i in range(len(string)):
    if string[i] == '*':
        count_mult += 1
    if string[i] == '+':
        count_plus += 1

print('Символ + встречается', count_plus, 'раз')
print('Символ * встречается', count_mult, 'раз')
```
