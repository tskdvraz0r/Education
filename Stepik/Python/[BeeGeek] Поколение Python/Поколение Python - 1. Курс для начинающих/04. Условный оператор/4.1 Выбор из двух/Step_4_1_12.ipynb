# Step 12 - Только + 🌶️

Напишите программу, которая считывает три числа и подсчитывает сумму только положительных чисел.

Формат входных данных: На вход программе подаются три целых числа.

Формат выходных данных: Программа должна вывести одно число – сумму положительных чисел.

Примечание. Если положительных чисел нет, то следует вывести 00.

## test:

Sample Input 1:

4

-22

1

Sample Output 1:

5

```python
# 08.09.2021

num1 = int(input())
num2 = int(input())
num3 = int(input())
if num1 < 0:
    num1 = 0
if num2 < 0:
    num2 = 0
if num3 < 0:
    num3 = 0
print(num1 + num2 + num3)
```
