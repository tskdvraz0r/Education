# Step 05 - Четное или нечетное?

Напишите программу, которая определяет, является число четным или нечетным.

Формат входных данных: На вход программе подаётся одно целое число.

Формат выходных данных: Программа должна вывести «Четное», если число четное, и «Нечетное» — если число нечетное.

## test:

Sample Input:

10

Sample Output:

Четное

```python
# 08.09.2021

num1= int(input())
if num1 % 2 == 0:
    print('Четное')
else:
    print('Нечетное')
```
