# Step 05 - Среднее число

Даны три различных целых числа. Напишите программу, которая находит среднее по величине число.

Формат входных данных: На вход программе подаётся три различных целых числа, каждое на отдельной строке.

Формат выходных данных: Программа должна вывести среднее число.

## test:

Sample Input:

1

2

3

Sample Output:

2

```python
# 10.09.2021

a, b, c = int(input()), int(input()), int(input())
if a > b > c or c > b > a:
    print(b)
elif b > c > a or a > c > b:
    print(c)
elif c > a > b or b > a > c:
    print(a)
```
