# Step 11 - Пересечение отрезков 🌶️🌶️

На числовой прямой даны два отрезка: [a(1); b(1)] и [a(2); b(2)]. Напишите программу, которая находит их пересечение.

Пересечением двух отрезков может быть:

 - отрезок;
 - точка;
 - пустое множество.
 
Формат входных данных: На вход программе подаются 4 целых числа a_1, b_1, a_2, b_2, каждое на отдельной строке. Гарантируется, что a_1 < b_1.

Формат выходных данных: Программа должна вывести на экран границы отрезка, являющегося пересечением, либо общую точку, либо текст «пустое множество».

## test:

Sample Input:

1

3

2

4

Sample Output:

2 3

```python
# 18.09.2021

a1, b1, a2, b2 = int(input()), int(input()), int(input()), int(input())
if a1 < b1 < a2 < b2:
    print('пустое множество')
elif a2 < b2 < a1 < b1:
    print('пустое множество')
elif a1 < b1 == a2 < b2:
    print(b1)
elif a2 < b2 == a1 < b1:
    print(b2)
elif a1 < a2 < b1 < b2:
    print(a2, b1)
elif a2 < a1 < b2 < b1:
    print(a1, b2)
elif a1 == a2 < b2 < b1:
    print(a2, b2)
elif a1 == a2 < b1 < b2:
    print(a2, b1)
elif a1 < a2 < b1 == b2:
    print(a2, b1)
elif a2 < a1 < b1 == b2:
    print(a1, b1)
elif a1 < a2 < b2 < b1:
    print(a2, b2)
elif a2 < a1 < b1 < b2:
    print(a1, b1)
elif a1 == a2 < b1 == b2:
    print(a1, b1)
```
