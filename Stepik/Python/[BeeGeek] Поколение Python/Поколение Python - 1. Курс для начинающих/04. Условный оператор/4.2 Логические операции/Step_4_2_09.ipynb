# Step 09 - Принадлежность 2

Напишите программу, которая принимает целое число xx и определяет, принадлежит ли данное число указанным промежуткам (-3]; [7)

Формат входных данных: На вход программе подаётся целое число xx.

Формат выходных данных: Программа должна вывести текст в соответствии с условием задачи.

## test:

Sample Input 1:

-44

Sample Output 1:

Принадлежит

```python
# 08.09.2021

num= int(input())

if not -3 < num < 7:
    print('Принадлежит')
else:
    print('Не принадлежит')
```
