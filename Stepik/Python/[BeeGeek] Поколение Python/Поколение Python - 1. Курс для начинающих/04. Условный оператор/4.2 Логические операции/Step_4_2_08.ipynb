# Step 08 - Принадлежность 1

Напишите программу, которая принимает целое число xx и определяет, принадлежит ли данное число указанному промежутку (-1; 17)

Формат входных данных: На вход программе подаётся целое число xx.

Формат выходных данных: Программа должна вывести текст в соответствии с условием задачи.

## test:

Sample Input:

2

Sample Output:

Принадлежит

```python
# 08.09.2021

num = int(input())
if (-1) < num <17:
    print('Принадлежит')
else:
    print('Не принадлежит')
```
