# Step 13 - Високосный год

Напишите программу, которая определяет, является ли год с данным номером високосным. Если год является високосным, то выведите «YES», иначе выведите «NO». Год является високосным, если его номер кратен 4, но не кратен 100, или если он кратен 400.

Формат входных данных: На вход программе подаётся натуральное число.

Формат выходных данных: Программа должна вывести текст в соответствии с условием задачи.

## test:

Sample Input:

2020

Sample Output:

YES

```python
# 09.09.2021

year= int(input())
if year % 4 ==0 and (year % 100 != 0) or year % 400 ==0:
    print('YES')
else:
    print('NO')
```

```python
# 24.11.2021

a = int(input())
print('YES' if a % 4 == 0 and a % 100 != 0 or a % 400 == 0 else 'NO')
```
