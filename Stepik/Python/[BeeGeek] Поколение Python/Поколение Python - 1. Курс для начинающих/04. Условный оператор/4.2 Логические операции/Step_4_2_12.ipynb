# Step 12 - Неравенство треугольника

Напишите программу, которая принимает три положительных числа и определяет, существует ли невырожденный треугольник с такими сторонами.

Формат входных данных: На вход программе подаётся три положительных целых числа.

Формат выходных данных: Программа должна вывести «YES» или «NO» в соответствии с условием задачи.

## test:

Sample Input:

5

2

3

Sample Output:

NO

```python
# 09.09.2021

a= int(input())
b= int(input())
c= int(input())

if a<b+c and b<c+a and c<a+b:
    print('YES')
else:
    print('NO')
```
