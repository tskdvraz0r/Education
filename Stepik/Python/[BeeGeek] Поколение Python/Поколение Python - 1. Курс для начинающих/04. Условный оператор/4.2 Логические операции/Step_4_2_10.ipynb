# Step 10 - Принадлежность 3

Напишите программу, которая принимает целое число xx и определяет, принадлежит ли данное число указанным промежуткам (-30;2], (7; 25].

Формат входных данных: На вход программе подаётся целое число xx.

Формат выходных данных: Программа должна вывести текст в соответствии с условием задачи.

## test:

Sample Input:

-332

Sample Output:

Не принадлежит

```python
# 08.09.2021

num= int(input())

if (-2 >= num > -30) or (7 < num <= 25):
    print('Принадлежит')
else:
    print('Не принадлежит')
```
