# Step 01 - Список четных

На вход программе подается четное число n, n ≥ 2. Напишите программу, которая выводит список четных чисел [2, 4, 6, ..., n].

**Формат входных данных:** На вход программе подается четное натуральное число.

**Формат выходных данных:** Программа должна вывести текст в соответствии с условием задачи.

## test: <br/>
Sample Input: <br/>
2

Sample Output: <br/>
[2]

```python
# 11.11.2021

print([i for i in range(int(input()) + 1) if i > 1 and i % 2 == 0])
```
