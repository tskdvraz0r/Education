# Step 04 - Две старушки

Две старушки идут навстречу друг другу с постоянными скоростями V_1 и V_2 км/ч. Определите, через какое время старушки встретятся, если расстояние между ними равно SS км.

Формат входных данных: На вход программе подаются три числа с плавающей точкой S, V_1, V_2, каждое на отдельной строке.

Формат выходных данных: Программа должна вывести одно число в соответствии с условием задачи.

## test:

Sample Input:

60.0

15.5

14.5

Sample Output:

2.0

```python
# 11.09.2021

s = float(input())  # Расстояние между бабушками
a = float(input())  # Скорость бабушки(1)
b = float(input())  # Скорость бабушки(2)
ab = a + b  # Общая скорость движения бабушек
print(round(s / ab, 2))
```
