# Step 03 - Площадь треугольника

Напишите программу, которая считывает длины двух катетов в прямоугольном треугольнике и выводит его площадь.

Формат входных данных: На вход программе подаётся два числа с плавающей точкой – длины катетов, каждое на отдельной строке.

Формат выходных данных: Программа должна вывести одно число – площадь треугольника.

## test:

Sample Input:

443

390

Sample Output:

86385.0

```python
# 11.09.2021

a = float(input())
b = float(input())
S = 1/2 * a * b
print(S)
```
