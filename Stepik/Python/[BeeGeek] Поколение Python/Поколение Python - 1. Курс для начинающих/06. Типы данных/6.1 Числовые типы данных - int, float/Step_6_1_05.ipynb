# Step 05 - Обратное число

Напишите программу, которая считывает с клавиатуры одно число и выводит обратное ему. Если при этом введённое с клавиатуры число – ноль, то вывести «Обратного числа не существует» (без кавычек).

Формат входных данных: На вход программе подается одно действительное число.

Формат выходных данных: Программа должна вывести действительное число обратное данному, либо текст в соответствии с условием задачи.

## test:

Sample Input:

2.5

Sample Output:

0.4

```python
# 11.09.2021

num = float(input())
if num == 0:
    print('Обратного числа не существует')
else:
    print(num ** - 1)
```
