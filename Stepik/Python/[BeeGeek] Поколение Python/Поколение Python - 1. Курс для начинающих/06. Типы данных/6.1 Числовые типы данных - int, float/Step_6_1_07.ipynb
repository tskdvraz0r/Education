# Step 07 - Dog age

На вход программе подается число nn – количество собачьих лет. Напишите программу, которая вычисляет возраст собаки в человеческих годах.

Формат входных данных: На вход программе подаётся натуральное число – количество собачьих лет.

Формат выходных данных: Программа должна вывести возраст собаки в человеческих годах.

Примечание. В течение первых двух лет собачий год равен 10.510.5 человеческим годам. После этого каждый год собаки равен 4 человеческим годам.

## test:

Sample Input:

1

Sample Output:

10.5

```python
# 11.09.2021

age = float(input())
if 0 < age <= 2:
    print(age * 10.5)
else:
    print((age - 2) * 4 + 21)
```
