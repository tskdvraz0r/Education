# Step 02 - Евклидово расстояние

На плоскости евклидово расстояние между двумя точками (x_1; y_1) и (x_2; y_2) определяется так: p = sqrt((x_1 - x_2) ** 2 + (y_1 - y_2) ** 2).

Напишите программу определяющую евклидово расстояние между двумя точками, координаты которых заданы.

Формат входных данных: На вход программе подается четыре вещественных числа, каждое на отдельной строке –  x_1, x_2, y_1, y_2.

Формат выходных данных: Программа должна вывести одно число – евклидово расстояние.

## test:

Sample Input:

2.0

2.5

44.155

100.50

Sample Output:

106.68197610187018

```python
# 18.09.2021

import math
x1, y1, x2, y2 = float(input()), float(input()), float(input()), float(input())
a = math.sqrt((x1 - x2)**2 + (y1 - y2)**2)
print(a)
```
