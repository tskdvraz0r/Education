# Step 03 - Напишите программу определяющую площадь круга и длину окружности по заданному радиусу R.

Формат входных данных: На вход программе подается одно вещественное число R​.

Формат выходных данных: Программа должна вывести два числа – площадь круга и длину окружности радиуса R.

## test:

Sample Input:

554.6

Sample Output:

966294.7126386268

3484.654571361799

```python
# 18.09.2021

import math
R = float(input())
S = math.pi*R**2
C = 2*math.pi*R
print(S)
print(C)
```
