# Step 12 - Цвет настроения синий

Напишите программу, которая считывает одну строку, после чего выводит «YES», если в введенной строке есть подстрока «синий» и «NO» в противном случае.

Формат входных данных: На вход программе подается одна строка.

Формат выходных данных: Программа должна вывести текст в соответствии с условием задачи.

## test:

Sample Input:

Какой хороший день!

Sample Output:

NO

```python
# 18.09.2021

s = input()
print('YES' if 'синий' in s else 'NO')
```
