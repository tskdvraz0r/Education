# Step 07 - Списочное выражение 1

На вход программе подается натуральное число n. Напишите программу, использующую списочное выражение, которая создает список содержащий квадраты чисел от 1 до n, а затем выводит его элементы построчно, то есть каждый на отдельной строке.

**Формат входных данных:** На вход программе подается натуральное число.

**Формат выходных данных:** Программа должна вывести текст в соответствии с условием задачи.

**Примечание:** Для вывода элементов списка используйте цикл for.

## test: <br/>
Sample Input: <br/>
5

Sample Output: <br/>
1 <br/>
4 <br/>
9 <br/>
16 <br/>
25 <br/>

```python
# 10.11.2021

print(*[i ** 2 for i in range(1, int(input()) + 1)], sep='\n')
```
