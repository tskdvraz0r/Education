# Step 12 - Удалите нечетные индексы

На вход программе подается натуральное число n, а затем n целых чисел. Напишите программу, которая создает из указанных чисел список, затем удаляет все элементы стоящие по нечетным индексам, а затем выводит полученный список.

**Формат входных данных:** На вход программе подаются натуральное число n, а затем n целых чисел, каждое на отдельной строке.

**Формат выходных данных:** Программа должна вывести список в соответствии с условием задачи.

**Примечание:** Используйте оператор del.

## test: <br/>
Sample Input: <br/>
10 <br/>
0 <br/>
1 <br/>
2 <br/>
3 <br/>
4 <br/>
5 <br/>
6 <br/>
7 <br/>
8 <br/>
9 <br/>

Sample Output: <br/>
[0, 2, 4, 6, 8]

```python
# 31.10.2021

n = int(input())
num_list = []

if n == 1:
    num = int(input())
    num_list.append(num)

else:
    for i in range(n):
        n = int(input())
        num_list.append(n)

    del num_list[1::2]

print(num_list)
```
