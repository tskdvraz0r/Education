# Step 09 - Список кубов

На вход программе подается натуральное число n, а затем n целых чисел. Напишите программу, которая создает из указанных чисел список их кубов.

**Формат входных данных:** На вход программе подаются натуральное число n, а затем n целых чисел, каждое на отдельной строке.

**Формат выходных данных:** Программа должна вывести список, состоящий из кубов указанных чисел.

## test: <br/>
Sample Input: <br/>
5 <br/>
1 <br/>
2 <br/>
3 <br/>
4 <br/>
5 <br/>

Sample Output: <br/>
[1, 8, 27, 64, 125]

```python
# 29.10.2021

mylist = []
for i in range(int(input())):
    mylist.append(int(input()) ** 3)
print(mylist)
```
