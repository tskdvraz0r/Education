# Step 07 - Список строк

На вход программе подается натуральное число n, а затем n строк. Напишите программу, которая создает из указанных строк список и выводит его.

**Формат входных данных:** На вход программе подаются натуральное число n, а затем n строк, каждая на отдельной строке.

**Формат выходных данных:** Программа должна вывести список состоящий из указанных строк.

## test: <br/>
Sample Input: <br/>
5 <br/>
C# <br/>
C++ <br/>
C <br/>
Python <br/>
F# <br/>

Sample Output: <br/>
['C#', 'C++', 'C', 'Python', 'F#']

```python
# 29.10.2021

mylist = list()
for i in range(int(input())):
    mylist.append(input())
print(mylist)
```
