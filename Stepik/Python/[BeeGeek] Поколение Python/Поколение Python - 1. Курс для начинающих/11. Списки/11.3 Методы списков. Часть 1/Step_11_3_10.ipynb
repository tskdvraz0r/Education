# Step 10 - Список делителей

На вход программе подается натуральное число n. Напишите программу, которая создает список состоящий из делителей введенного числа.

**Формат входных данных:** На вход программе подается натуральное число n.

**Формат выходных данных:** Программа должна вывести список, состоящий из делителей введенного числа.

## test: <br/>
Sample Input: <br/>
17

Sample Output: <br/>
[1, 17]

```python
# 29.10.2021

n = int(input())
mylist = list()

for i in range(1, n + 1):
    if n % i == 0:
        mylist.append(i)

print(mylist)
```
