Напишите программу, выводящую следующий список: ['a', 'bb', 'ccc', 'dddd', 'eeeee', 'ffffff', ...]

**Формат выходных данных:** Программа должна вывести указанный список.

**Примечание:** Последний элемент списка состоит из 26 символов z.

```python
# 29.10.2021

abc = list()
for i in range(26):
    abc.append(chr(i + 97) * (i + 1))

print(abc)
```
