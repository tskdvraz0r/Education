# Step 06 - Диаграмма

На вход программе подается строка текста, содержащая целые числа. Напишите программу, которая по заданным числам строит столбчатую диаграмму.

**Формат входных данных:** На вход программе подается строка текста, содержащая целые числа, разделенных символом пробела.

**Формат выходных данных:** Программа должна вывести столбчатую диаграмму.

## test: <br/>
Sample Input: <br/>
5 3 1 7 10 2
```
Sample Output:
+++++
+++
+
+++++++
++++++++++
++

````python
# 05.11.2021

n = input()
n = n.split()

for i in range(len(n)):
    print('+' * int(n[i]))
````
