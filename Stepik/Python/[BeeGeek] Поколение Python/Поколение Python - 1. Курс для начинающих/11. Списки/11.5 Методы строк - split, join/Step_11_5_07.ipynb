# Step 07 - Корректный ip-адрес

На вход программе подается строка текста, содержащая 4 целых числа разделенных точкой. Напишите программу, которая определяет является ли введенная строка текста корректным ip-адресом.

**Формат входных данных:** На вход программе подается строка текста, содержащая 4 целых числа разделенных точкой.

**Формат выходных данных:** Программа должна вывести «ДА», если введеная строка является корректным ip-адресом, и «НЕТ» — в противном случае.

**Примечание:** ip-адрес является корректным, если все 4 числа находятся в диапазоне от 0 до 255 включительно.

## test: <br/>
Sample Input: <br/>
192.168.0.3

Sample Output: <br/>
ДА

```python
# 05.11.2021

ip = input()
ip = ip.split('.')
flag = True

if len(ip) == 4:
    for i in range(4):
        if int(ip[i]) < 0 or int(ip[i]) > 255:
            flag = False
            
if flag == True:
    print('ДА')
else:
    print('НЕТ')
```
