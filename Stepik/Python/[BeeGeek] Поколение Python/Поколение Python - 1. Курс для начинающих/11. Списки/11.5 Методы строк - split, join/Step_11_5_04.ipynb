# Step 04 - Инициалы

На вход программе подается строка текста, содержащая имя, отчество и фамилию человека. Напишите программу, которая выводит инициалы человека.

**Формат входных данных:** На вход программе подается строка текста, содержащая имя, отчество и фамилию человека.

**Формат выходных данных:** Программа должна вывести текст в соответствии с условием задачи.

## test: <br/>
Sample Input: <br/>
Владимир Семенович Высоцкий

Sample Output: <br/>
В.С.В.

```python
# 05.11.2021

s = input()
fio = []
s = s.split()

for i in range(len(s)):
    fio.append(s[i][0])

print('.'.join(fio), end='.', sep='')
```
