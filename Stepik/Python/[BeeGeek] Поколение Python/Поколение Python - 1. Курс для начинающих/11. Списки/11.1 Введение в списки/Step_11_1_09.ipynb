# Step 09 - Список букв

На вход программе подается одно число n. Напишите программу, которая выводит список, состоящий из n букв английского алфавита ['a', 'b', 'c', ...] в нижнем регистре.

**Формат входных данных:** На вход программе подается натуральное число n, n ≤ 26.

**Формат выходных данных:** Программа должна вывести текст в соответствии с условием задачи.

## test: <br/>
Sample Input: <br/>
5

Sample Output: <br/>
['a', 'b', 'c', 'd', 'e']

```python
# 29.10.2021

n = int(input())
abc = list()

if n <= 26:
    for i in range(n):
        abc += chr(i + 97)

print(abc)
```
