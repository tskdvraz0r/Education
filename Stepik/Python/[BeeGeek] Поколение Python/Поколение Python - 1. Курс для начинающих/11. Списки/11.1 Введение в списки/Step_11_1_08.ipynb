# Step 08 - Список чисел

На вход программе подается одно число n. Напишите программу, которая выводит список [1, 2, 3, ..., n].

**Формат входных данных:** На вход программе подается одно натуральное число.

**Формат выходных данных:** Программа должна вывести текст в соответствии с условием задачи.

## test: <br/>
Sample Input: <br/>
5

Sample Output: <br/>
[1, 2, 3, 4, 5]

```python
# 29.10.2021

print(list(range(1, int(input()) + 1)))
```
