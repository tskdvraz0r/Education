Дополните приведенный код, так чтобы он вывел сумму квадратов элементов списка numbers.

```python
# 31.10.2021

numbers = [1, 78, 23, -65, 99, 9089, 34, -32, 0, -67, 1, 11, 111]
num_sum = 0
total_sum = 0

for i in numbers:
    num_sum = i ** 2
    total_sum += num_sum

print(total_sum)
```
