# Step 03 - Значение функции

На вход программе подается натуральное число n, а затем n целых чисел. Напишите программу, которая для каждого введенного числаx выводит значение функции: <br/>
f(x) = x ** 2 + 2x + 1, каждое на отдельной строке.

**Формат входных данных:** На вход программе подаются натуральное число n, а затем n целых чисел, каждое на отдельной строке.

**Формат выходных данных:** Программа должна вывести сначала введенные числа, затем пустую строку, а затем соответствующие значения функции.

**Примечание:** Для первого теста имеем: <br/>
f(1) = 1 ** 2 + 2 * 1 + 1 = 4 <br/>
f(2) = 2 ** 2 + 2 * 2 + 1 = 9 <br/>
f(3) = 3 ** 2 + 2 * 3 + 1 = 16 <br/>
...

## test: <br/>
Sample Input: <br/>
5 <br/>
1 <br/>
2 <br/>
3 <br/>
4 <br/>
5 <br/>

Sample Output: <br/>
1 <br/>
2 <br/>
3 <br/>
4 <br/>
5 <br/>
... <br/>
4 <br/>
9 <br/>
16 <br/>
25 <br/>
36 <br/>

```python
# 31.10.2021

n = int(input())
num_list = []

for i in range(n):
    num = int(input())
    num_list.append(num)

print(*num_list, sep='\n')
print()

for j in num_list:
    f = j ** 2 + 2 * j + 1
    print(f)
```
