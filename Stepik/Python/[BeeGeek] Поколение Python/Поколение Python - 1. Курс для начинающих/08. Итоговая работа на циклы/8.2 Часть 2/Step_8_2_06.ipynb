# Step 06 - Все вместе 2

Дано натуральное число. Напишите программу, которая вычисляет:
- количество цифр 3 в нем;
- сколько раз в нем встречается последняя цифра;
- количество четных цифр;
- сумму его цифр, больших пяти;
- произведение цифр, больших семи (если цифр больших семи нет, то вывести 1, если такая цифра одна, то вывести ее);
- сколько раз в нем встречается цифры 0 и 5 (всего суммарно).

Формат входных данных: На вход программе подается одно натуральное число.

Формат выходных данных: Программа должна вывести значения указанных величин в указанном порядке.

## test: <br/>
Sample Input: <br/>
56639

Sample Output: <br/>
1 <br/>
1 <br/>
2 <br/>
21 <br/>
9 <br/>
1 <br/>

```python
# 27.10.2021

num = int(input())
last_digital = num % 10
count_3 = 0
count_ld = 0
count_del2 = 0
sum_digital_5 = 0
mult_digital_7 = 1
count_digital_05 = 0

while num != 0:

    # считаем количество цифр 3
    if num % 10 == 3:
        count_3 += 1
    # считаем сколько раз встречается последняя цифра
    if num % 10 == last_digital:
        count_ld += 1
    # считаем количество чётных цифр
    if num % 2 == 0:
        count_del2 += 1
    # считаем сумму цифр больше 5
    if num % 10 > 5:
        num5 = num % 10
        sum_digital_5 += num5
    # считаем сумму цифр больше 7
    if num % 10 > 7:
        num7 = num % 10
        mult_digital_7 *= num7
    #
    if num % 10 == 5 or num % 10 == 0:
        num05 = num % 10
        count_digital_05 += 1
    # отсекаем последюю цифру
    num //= 10

print(count_3, count_ld, count_del2, sum_digital_5, mult_digital_7, count_digital_05, sep='\n')
```
