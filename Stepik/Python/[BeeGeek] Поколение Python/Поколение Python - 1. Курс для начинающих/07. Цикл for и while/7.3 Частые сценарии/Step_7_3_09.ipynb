# Step 09 - Факториал

На вход программе подается натуральное число n. Напишите программу, которая вычисляет n!.

Входные данные: На вход программе подается натуральное число n,(n ≤ 12).

Выходные данные: Программа должна вывести единственное число в соответствии с условием задачи.

Примечание. Факториалом натурального числа n, называется произведение всех натуральных чисел от 1 до n, то есть n! = 1 * 2 * 3 * ... n

## test:

Sample Input:

3

Sample Output:

6

```python
# 02.10.2021

n = int(input())
total = 1
for i in range(1, n + 1):
    total *= i
print(total)
```
