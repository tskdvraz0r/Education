# Step 12 - Знакочередующаяся сумма

На вход программе подается натуральное число n. Напишите программу вычисления знакочередующей суммы 1 − 2 + 3 - 4 + 5 - 6 + ... + (-1) ** (n + 1)n

Входные данные: На вход программе подается натуральное число n.

Выходные данные: Программа должна вывести единственное число в соответствии с условием задачи.

## test:

Sample Input:

5

Sample Output:

3

```python
# 02.10.2021

total = 0
n = int(input())
for i in range(n + 1):
    if (i+1) % 2 != 0:
        total -= i
    else:
        total += i
print(total)
```
