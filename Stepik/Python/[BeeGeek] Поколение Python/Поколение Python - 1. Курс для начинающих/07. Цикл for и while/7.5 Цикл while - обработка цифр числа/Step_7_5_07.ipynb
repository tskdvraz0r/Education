# Step 07 - Все вместе

Дано натуральное число. Напишите программу, которая вычисляет:
- сумму его цифр;
- количество цифр в нем;
- произведение его цифр;
- среднее арифметическое его цифр;
- его первую цифру;
- сумму его первой и последней цифры.
 
Формат входных данных: На вход программе подается одно натуральное число.

Формат выходных данных: Программа должна вывести значения указанных величин в указанном порядке.

## test:

Sample Input:

5678

Sample Output:

26

4

1680

6.5

5

13

```python
# 25.10.2021

num = int(input())
sum_num = 0
cnt_num = 0
pro_num = 1
last_num = num % 10
revers = str()

while num != 0:
    last_digital = num % 10
    sum_num += last_digital
    cnt_num += 1
    pro_num *= last_digital
    revers = revers + str(last_digital)
    num = num // 10

print(sum_num)  # сумма
print(cnt_num)  # количество
print(pro_num)  # произведение
print(sum_num / cnt_num)  # среднее
print(int(revers) % 10)  # первая
print(last_num + (int(revers) % 10))  # сумма первой и последней
```
