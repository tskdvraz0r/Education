# Step 09 - Одинаковые цифры

Дано натуральное число. Напишите программу, которая определяет, состоит ли указанное число из одинаковых цифр.

Формат входных данных: На вход программе подается одно натуральное число.

Формат выходных данных: Программа должна вывести «YES» если число состоит из одинаковых цифр и «NO» в противном случае.

## test:

Sample Input:

11111

Sample Output:

YES

```python
# 25.10.2021

num = int(input())  # enter num
last_num = num % 10  # last number
same_num = False  # flag

while num != 0:
    last_digital = num % 10
    if last_digital == last_num:
        same_num = True
    else:
        same_num = False
        break

    last_num = last_digital
    num = num // 10

if same_num == True:
    print('YES')
else:
    print('NO')
```
