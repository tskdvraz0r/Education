# Step 09 - Звездный треугольник

На вход программе подается натуральное число n(n≥2) – катет прямоугольного равнобедренного треугольника.

Напишите программу, которая выводит звездный треугольник в соответствии с примером.

Формат входных данных: На вход программе подается одно натуральное число n(n≥2).

Формат выходных данных: Программа должна вывести треугольник в соответствии с условием задачи.

## test:

Sample Input:

3

Sample Output:

'***'

'**'

'*'

```python
# 01.10.2021

a = int(input())
for i in range(a):
    print((a - i) * '*')
```
