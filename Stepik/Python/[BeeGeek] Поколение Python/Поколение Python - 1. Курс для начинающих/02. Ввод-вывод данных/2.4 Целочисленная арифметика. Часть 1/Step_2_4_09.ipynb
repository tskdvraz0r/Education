# Step 09 - Следующее и предыдущее

Напишите программу, которая считывает целое число, после чего на экран выводится следующее и предыдущее целое число с пояснительным текстом.

Формат входных данных: На вход программе подаётся целое число.

Формат выходных данных: Программа должна вывести текст согласно условию задачи.

## test:

Sample Input:

20

Sample Output:

Следующее за числом 20 число: 21

Для числа 20 предыдущее число: 19

```python
# 06.09.2021

a = int(input())
b = a - 1
c = a + 1
print('Следующее за числом', a, 'число:', c)
print('Для числа', a, 'предыдущее число:', b)
```

```python
# 07.07.2022

number = int(input())
print(
    f"Следующее за числом {number} число: {number + 1}",
    f"Для числа {number} предыдущее число: {number - 1}",
    sep = "\n"
)
```
