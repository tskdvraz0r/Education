# Step 06 - Сумма трёх чисел

Напишите программу, которая считывает три целых числа и выводит на экран их сумму. Каждое число записано в отдельной строке.

Формат входных данных: На вход программе подаётся три целых числа, каждое на отдельной строке.

Формат выходных данных: Программа должна вывести сумму введенных чисел.

## test:

Sample Input:

9

11

2

Sample Output:

22

```python
# 05.09.2021

x = int(input())
y = int(input())
z = int(input())
print(x+y+z)
```
