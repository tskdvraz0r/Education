# Step 02 - Сумма квадратов VS квадрат суммы

Напишите программу, которая считывает два целых числа a и b и выводит на экран квадрат суммы (a + b) ** 2 и сумму квадратов a ** 2 + b ** 2 этих чисел.

Формат входных данных: На вход программе подаётся два целых числа, каждое на отдельной строке.

Формат выходных данных: Программа должна вывести текст в соответствии с условием.

## test:

Sample Input:

3

2

Sample Output:

Квадрат суммы 3 и 2 равен 25

Сумма квадратов 3 и 2 равна 13

```python
# 06.09.2021

a=int(input())
b=int(input())
print('Квадрат суммы', a, 'и', b, 'равен', (a+b)**2)
print('Сумма квадратов', a, 'и', b, 'равна', (a**2+b**2))
```

```python
# 08.07.2022

a, b = int(input()), int(input())
print(
    f"Квадрат суммы {a} и {b} равен {(a + b) ** 2}",
    f"Сумма квадратов {a} и {b} равен {(a ** 2 + b ** 2)}",
    sep = "\n"
)
```
