```python
def square_of_tree(hight: int) -> int:
    """
    Notes:
        Функция принимает на вход "высоту" ёлки и возвращает "площадь".
        
         #       #       #
        ###     ###     ###
         #       #       #
               #####   ##### 
                 #       #
                      #######
                         #
    
    Args:
        hight (int): Высота ёлки.
    
    Raises:
        ValueError: "Некорректный тип данных;"
    
    Returns:
        (int): Площадь ёлки.
    
    Samples:
        >>> square_of_tree(hight=0)
        <<< 1
        
        >>> square_of_tree(hight=1)
        <<< 5
        
        >>> square_of_tree(hight=2)
        <<< 11
    """
    
    logger.debug(f"INPUT: {hight=};")
    
    # Проверка входных данных;
    Check.value_type(
        value=hight,
        expected_type=int
    )
    
    # Реализация алгоритма;
    square: float = hight * (hight + 3) + 1
    logger.debug(f"OUTPUT: {int(square)};")
    
    return int(square)


if __name__ == "__main__":
    print(square_of_tree(hight=5))
```

```python
def cinema(cnt_boys_and_girls: str) -> str:
    """
    Notes:
        Функция принимает на вход строку с двумя целыми числами (количество парней и девушек, 
        которые вошли в кинотеатр). Возвращает строку вида "BGBBG...", как им сесть так, чтобы у 
        каждого был хотябы один сосед противоположного пола. 

    Args:
        cnt_boys_and_girls (str): Строка с целыми числами.
    
    Raises:
        ValueError: "Некорректный тип данных;"
        ValueError: "Передано нечисловое значение;"

    Returns:
        str: Строка с "посадкой".
    
    Samples:
        >>> cinema(cnt_boys_and_girls="5 3")
        <<< "BGBBGBBG"
    """
    
    logger.debug(f"INPUT: {cnt_boys_and_girls=}")
    
    # Проверка входных данных;
    Check.value_type(
        value=cnt_boys_and_girls,
        expected_type=str
    )
    
    # Получение значений;
    try:
        cnt_boys, cnt_girls = map(int, cnt_boys_and_girls.split())
    
    except ValueError as err:
        logger.critical("Передано нечисловое значение;")
        sys.exit(1)
    
    
    # Реализация алгоритма;
    list_boys: list[str] = ["B"] * cnt_boys
    list_girls: list[str] = ["G"] * cnt_girls
    result: str = ""
    
    if cnt_boys < ((cnt_girls + 1) // 2) or cnt_girls < ((cnt_boys + 1) // 2):
        print("NO SOLUTION")

    elif cnt_boys == cnt_girls:
        print("BG" * cnt_boys)

    else:
        for _ in range(cnt_boys + cnt_girls):
            len_boys: int = len(list_boys)
            len_girls: int = len(list_girls)
            
            if len(result) == 1 and result.startswith("B"):
                result += list_girls.pop()
                len_girls = len(list_girls)
            
            elif len(result) == 1 and result.startswith("G"):
                result += list_boys.pop()
                len_boys = len(list_boys)
            
            elif len_boys > (len_girls // 2) and not result.endswith("BB"):
                result += list_boys.pop()
                len_boys = len(list_boys)

            elif len_girls > (len_boys // 2) and not result.endswith("GG"):
                result += list_girls.pop()
                len_girls = len(list_girls)
    
    logger.debug(f"OUTPUT: {result=}")
    
    return result


if __name__ == "__main__":
    print(
        cinema(cnt_boys_and_girls="15 13")
    )
```

```python
def divide_competition(k: int, m: int, a: int, b: int) -> int:
    """
    Notes:
        Функция принимает 4 целых числа: k=число Кати, m=число Маши, a-b=диапазон чисел включительно.
        Возвращает разницу количества чисел, которые делятся на "заданные" (k, m).

    Args:
        k (int): Любимое число Кати.
        m (int): Любимое число Маши.
        a (int): Нижняя граница диапазона (включительно).
        b (int): Верхняя граница диапазона (включительно).
    
    Raises:
        ValueError: "Некорректный тип данных;"

    Returns:
        int: Разница между количствами чисел, делящихся на "заданные" (k, m).
    
    Samples:
        >>> divide_competition(k=2, m=3, a=2, b=9)
        <<< 1
        
        >>> divide_competition(k=3, m=3, a=6, b=6)
        <<< 0
    """
    
    logger.debug(f"INPUT: {k=};")
    logger.debug(f"INPUT: {m=};")
    logger.debug(f"INPUT: {a=};")
    logger.debug(f"INPUT: {b=};")
    
    # Проверка входных данных;
    for value in (k, m, a, b):
        Check.value_type(
            value=value,
            expected_type=int
        )
    
    
    # Реализация алгоритма;
    result: int = (b // k - (a - 1) // k) - (b // m - (a - 1) // m)
    
    logger.debug(f"OUTPUT: {result=};")
    
    return result


if __name__ == "__main__":
    print(
        divide_competition(k=2, m=3, a=3, b=20)
    )
```
